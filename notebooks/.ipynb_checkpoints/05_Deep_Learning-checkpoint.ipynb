```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

Path('outputs').mkdir(exist_ok=True)

# Escolha FAST=True para rodar mais rápido (usa últimos 3 anos)
FAST = True

# Funções de métricas (robustas)
def mae(y_true, y_pred):
    y_true, y_pred = pd.Series(y_true), pd.Series(y_pred)
    dfm = pd.concat([y_true, y_pred], axis=1).dropna()
    return (dfm.iloc[:,0] - dfm.iloc[:,1]).abs().mean()

def rmse(y_true, y_pred):
    y_true, y_pred = pd.Series(y_true), pd.Series(y_pred)
    dfm = pd.concat([y_true, y_pred], axis=1).dropna()
    return np.sqrt(((dfm.iloc[:,0] - dfm.iloc[:,1])**2).mean())

def mape(y_true, y_pred):
    y_true, y_pred = pd.Series(y_true), pd.Series(y_pred)
    dfm = pd.concat([y_true, y_pred], axis=1).dropna()
    dfm = dfm[dfm.iloc[:,0] != 0]
    return (dfm.iloc[:,0].sub(dfm.iloc[:,1]).abs().div(dfm.iloc[:,0]).mean()) * 100

print('Config OK. FAST =', FAST)
```

```python
csv_path = Path('data/btc_limpo.csv')

if csv_path.exists():
    df = pd.read_csv(csv_path, parse_dates=['Date'], index_col='Date')
    print('Carregado de', csv_path)
else:
    import yfinance as yf
    df = yf.download('BTC-USD', start='2016-01-01', end=None)
    print('Baixado do yfinance')

# achatar MultiIndex de colunas
if isinstance(df.columns, pd.MultiIndex):
    df.columns = ['_'.join([str(a) for a in col if str(a)]) for col in df.columns]
df.columns = [c.strip() for c in df.columns]

# tentar detectar coluna de fechamento
if 'Close' in df.columns:
    close_col = 'Close'
else:
    possibles = [c for c in df.columns if 'close' in c.lower()]
    if possibles:
        close_col = possibles[0]
    else:
        raise ValueError('Coluna de fechamento (Close) não encontrada. Verifique o CSV.')

# manter apenas Close; opcionalmente manter Volume se existir
df = df[[close_col] + ([c for c in ['Volume','volume'] if c in df.columns])].copy()
df = df.rename(columns={close_col:'Close'})

# garantir numérico, índice datetime, ordenar, definir frequência diária e preencher
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
df = df.dropna(subset=['Close']).sort_index()
df = df.asfreq('D')
df['Close'] = df['Close'].ffill()

# opcional FAST: reduzir para últimos 3 anos
if FAST:
    cutoff = df.index.max() - pd.Timedelta(days=3*365)
    df = df[df.index >= cutoff].copy()

print('Dados prontos:', df.index.min(), '→', df.index.max(), '| pontos:', len(df))
df.head()
```

```python
series = df['Close'].copy()
split = int(len(series) * 0.8)
train = series.iloc[:split].copy()
test  = series.iloc[split:].copy()
print('Split feito — treino:', len(train), 'teste:', len(test))
```

```python
pred_naive = series.shift(1).loc[test.index]
pred_snaive7 = series.shift(7).loc[test.index]

df_preds = pd.DataFrame({
    'Real': test,
    'Naive': pred_naive,
    'SNaive7': pred_snaive7
})
df_preds.to_csv('outputs/predictions_baselines.csv')
print('Baselines salvos: outputs/predictions_baselines.csv')
display(df_preds.head())
```

```python
from statsmodels.tsa.arima.model import ARIMA

order = (1,1,1)  # ponto de partida; você pode alterar
model = ARIMA(train, order=order)
res = model.fit()
pred_arima_static = res.forecast(steps=len(test))
pred_arima_static.index = test.index

df_preds['ARIMA_static'] = pred_arima_static
df_preds.to_csv('outputs/predictions_with_arima_static.csv')
print('ARIMA (estático) salvo: outputs/predictions_with_arima_static.csv')
display(df_preds.tail())
```

```python
def plot_zoom(series, preds_df, fname, dias=730):
    end = series.index.max()
    start = end - pd.Timedelta(days=dias)
    plt.figure(figsize=(12,6))
    plt.plot(series[start:].index, series[start:].values, label='Real', linewidth=1.2)
    for col in preds_df.columns:
        p = preds_df[col].reindex(series[start:].index)
        if p.isna().all():
            continue
        plt.plot(p.index, p.values, label=col, linewidth=1)
    plt.title(fname.replace('_',' '))
    plt.legend()
    plt.tight_layout()
    plt.savefig('outputs/' + fname, dpi=150)
    plt.show()

# ARIMA static vs Real (2 anos)
plot_zoom(series, df_preds[['ARIMA_static']], 'plot_ARIMA_static_vs_real_2y.png', dias=730)

# Baselines vs Real (2 anos)
plot_zoom(series, df_preds[['Naive','SNaive7']], 'plot_baselines_vs_real_2y.png', dias=730)

# ML comparativo (vai ser gerado depois) — reserve espaço
print('PNGs gerados em outputs/')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 1) Achatar colunas do MultiIndex
df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]

# 2) Usar apenas o fechamento (Close)
# Aqui eu assumo que sua coluna ficou com o nome 'Close_BTC-USD'
close = df['Close']

# Divide treino e teste (80/20 por exemplo)
n = int(len(close) * 0.8)
train, test = close[:n], close[n:]

# 3) ARIMA com walk-forward
history = list(train)
predictions = []

for t in range(len(test)):
    model = ARIMA(history, order=(1,1,1))
    model_fit = model.fit()
    yhat = model_fit.forecast()[0]
    predictions.append(yhat)
    history.append(test.iloc[t])

# Converte previsões em Series com as mesmas datas do teste
pred_series = pd.Series(predictions, index=test.index)

# 4) Plot – apenas últimos 2 anos
plt.figure(figsize=(12,6))
plt.plot(close[-730:], label='Real', color='black')
plt.plot(pred_series[-730:], label='ARIMA (walk-forward)', color='red')
plt.title('Bitcoin – Previsão ARIMA vs Real (últimos 2 anos)')
plt.xlabel('Data')
plt.ylabel('Preço (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
try:
    from prophet import Prophet
except Exception:
    import subprocess
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'prophet'])
    from prophet import Prophet

# preparar dataframe para Prophet
prophet_df = series.reset_index().rename(columns={'Date':'ds','Close':'y'}) if 'Date' in series.reset_index().columns else series.reset_index().rename(columns={series.index.name:'ds', series.name:'y'})
prophet_df['ds'] = pd.to_datetime(prophet_df['ds'])
train_prop = prophet_df.iloc[:split].copy()
test_prop = prophet_df.iloc[split:].copy()

m = Prophet(daily_seasonality=False, weekly_seasonality=True, yearly_seasonality=True)
m.fit(train_prop)
future = m.make_future_dataframe(periods=len(test_prop), freq='D')
fc = m.predict(future)
fc_test = fc.set_index('ds').loc[test_prop['ds'], 'yhat']
fc_test.index = test_prop['ds'].values
df_preds['Prophet'] = fc_test
df_preds.to_csv('outputs/predictions_with_prophet.csv')
print('Prophet pronto e salvo: outputs/predictions_with_prophet.csv')
```

```python
rows = []
models_now = [c for c in ['Naive','SNaive7','ARIMA_static','ARIMA_rolling','Prophet'] if c in df_preds.columns]
for name in models_now:
    rows.append({
        'model': name,
        'MAE': mae(df_preds['Real'], df_preds[name]),
        'RMSE': rmse(df_preds['Real'], df_preds[name]),
        'MAPE(%)': mape(df_preds['Real'], df_preds[name])
    })
metrics_all = pd.DataFrame(rows).set_index('model').round(6)
metrics_all.to_csv('outputs/resultados_all_models.csv')
display(metrics_all)
print('Métricas completas salvas em outputs/resultados_all_models.csv')
```

```python
# partir de df (que já tem Close)
Xdf = df.copy()

# lags
for lag in [1,2,3,7,14]:
    Xdf[f'lag_close_{lag}'] = Xdf['Close'].shift(lag)

# returns
Xdf['return_1'] = Xdf['Close'].pct_change(1)
Xdf['return_7'] = Xdf['Close'].pct_change(7)

# SMA/EMA
Xdf['sma_7'] = Xdf['Close'].rolling(7).mean()
Xdf['sma_20'] = Xdf['Close'].rolling(20).mean()
Xdf['ema_12'] = Xdf['Close'].ewm(span=12, adjust=False).mean()
Xdf['ema_26'] = Xdf['Close'].ewm(span=26, adjust=False).mean()

# MACD
Xdf['macd'] = Xdf['ema_12'] - Xdf['ema_26']
Xdf['macd_signal'] = Xdf['macd'].ewm(span=9, adjust=False).mean()

# RSI (14)
delta = Xdf['Close'].diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)
avg_gain = gain.ewm(alpha=1/14, min_periods=14).mean()
avg_loss = loss.ewm(alpha=1/14, min_periods=14).mean()
rs = avg_gain / avg_loss
Xdf['rsi_14'] = 100 - (100 / (1 + rs))

# vol
Xdf['vol_30'] = Xdf['Close'].pct_change().rolling(30).std()

# target: próximo dia close (t+1)
Xdf['target_close_1'] = Xdf['Close'].shift(-1)

# limpar NaNs
Xdf = Xdf.dropna()
print('Features prontas, shape:', Xdf.shape)
```

```python
split_ml = int(len(Xdf) * 0.8)
train_ml = Xdf.iloc[:split_ml].copy()
test_ml  = Xdf.iloc[split_ml:].copy()

features = [c for c in Xdf.columns if c.startswith('lag_') or c in ['return_1','return_7','sma_7','sma_20','ema_12','ema_26','macd','macd_signal','rsi_14','vol_30']]
X_train = train_ml[features]
y_train = train_ml['target_close_1']
X_test  = test_ml[features]
y_test  = test_ml['target_close_1']

print('Train ML:', X_train.shape, 'Test ML:', X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print('LinearRegression — MAE:', mean_absolute_error(y_test, y_pred_lr),
      ' RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_lr)),
      ' MAPE(%):', (np.mean(np.abs((y_test - y_pred_lr)/y_test)) * 100))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print('RandomForest — MAE:', mean_absolute_error(y_test, y_pred_rf),
      ' RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_rf)),
      ' MAPE(%):', (np.mean(np.abs((y_test - y_pred_rf)/y_test)) * 100))
```

```python
out_ml = pd.DataFrame(index=test_ml.index)
out_ml['Real'] = y_test
out_ml['LR'] = y_pred_lr
out_ml['RF'] = y_pred_rf
out_ml.to_csv('outputs/predictions_ml.csv')

metrics_ml = pd.DataFrame({
    'model': ['LinearRegression','RandomForest'],
    'MAE': [mean_absolute_error(y_test, y_pred_lr), mean_absolute_error(y_test, y_pred_rf)],
    'RMSE': [np.sqrt(mean_squared_error(y_test, y_pred_lr)), np.sqrt(mean_squared_error(y_test, y_pred_rf))],
    'MAPE(%)': [np.mean(np.abs((y_test - y_pred_lr)/y_test))*100, np.mean(np.abs((y_test - y_pred_rf)/y_test))*100]
})
metrics_ml.to_csv('outputs/resultados_ml.csv', index=False)
print('ML outputs salvos em outputs/ (predictions_ml.csv e resultados_ml.csv)')

# plot comparativo (último ano)
start = out_ml.index.max() - pd.Timedelta(days=365)
plt.figure(figsize=(12,6))
for c in out_ml.columns:
    plt.plot(out_ml[start:].index, out_ml[start:][c], label=c, linewidth=1)
plt.legend(); plt.title('Comparativo ML (último ano)'); plt.tight_layout()
plt.savefig('outputs/plot_ml_compare_1y.png', dpi=150)
plt.show()
```
